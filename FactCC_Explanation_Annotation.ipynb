# Explanation Analysis

### Example data use

```python
import json, random

with open('data/factcc/factcc_explanation_annotation.json', 'r') as f:
    data = json.load(f)

# get a single sample
sample = random.choice(data)

print("[Input FactCC Document] %s" % sample["doc"])
print("------")
print("[Known inconsistent summary] %s" % sample["summary"])
print("------")

anno = random.choice(sample["samples"])
print("[%s Explanation] %s" % (anno["model"], anno["explanation"]))
print("[Annotation of Explanation Type] %s" % (anno["annotation"]))
```
